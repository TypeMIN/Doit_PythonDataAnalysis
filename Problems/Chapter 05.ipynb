## Analysis Challenge

```python
mw = pd.read_csv('midwest.csv')
```

```python
mw.info()
```

```python
mw.describe(include = 'all')
```

```python
mw.head(10)
```

```python
mw = mw.rename(columns = {'poptotal' : 'total', 'popasian' : 'asian'})
mw.head()
```

```python
mw['perasian'] = mw['asian'] / mw['total']
mw['perasian'].plot.hist()
```

```python
mw['perasian'].mean()
```

```python
import numpy as np
mw['isoverasian'] = np.where(mw['perasian'] > mw['perasian'].mean(), 'large', 'small')
mw.head()
```

```python
is_overasian = mw['isoverasian'].value_counts()
is_overasian.plot.bar(rot = 0)
```
